```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train,y_train)
y_pred_train2 = dtr.predict(X_train)
y_pred_test2 = dtr.predict(X_test)
from sklearn.metrics import r2_score
print("Training Accuracy: ",r2_score(y_train, y_pred_train2))
print("\n\nTesting Accuracy: ",r2_score(y_test, y_pred_test2))
```

```python
dtr2 = DecisionTreeRegressor(max_depth= 7, max_features= 'auto',max_leaf_nodes= 10, min_samples_leaf= 5,min_weight_fraction_leaf=0.1,splitter= 'best')
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=200,max_depth = 10)
rf.fit(X_train,y_train)
y_pred_train3 = rf.predict(X_train)
y_pred_test3 = rf.predict(X_test)
print("Training Accuracy: ",r2_score(y_train, y_pred_train3))
print("\n\nTesting Accuracy: ",r2_score(y_test, y_pred_test3))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
step1 = ColumnTransformer(transformers=[
                       ('col_tnf',OneHotEncoder(sparse = False,drop = 'first'),[0,1,3,9,13])         
],remainder = 'passthrough')
step2 = RandomForestRegressor(n_estimators = 100,random_state = 3,max_depth = 15)
pipe = Pipeline([
       ('step1',step1),
       ('step2',step2)          
])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('r2',r2_score(y_test,y_pred))

    
```
